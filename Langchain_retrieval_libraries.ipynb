```python
import os

os.environ['CHROMEDRIVER_PATH'] = '/usr/lib/chromium-browser/chromedriver'
os.environ['CHROME_BINARY_PATH'] = '/usr/bin/chromium-browser'
```

```python
from langchain.document_loaders import SeleniumURLLoader
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Configure Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.binary_location = os.environ['CHROME_BINARY_PATH']

urls = [
  "https://www.youtube.com/watch?v=TFa539R09EQ&t=139s",
  "https://www.youtube.com/watch?v=6Zv6A_9urh4&t=112s"
]

# Pass the configured driver to SeleniumURLLoader
loader = SeleniumURLLoader(urls=urls, selenium_options=chrome_options, executable_path=os.environ['CHROMEDRIVER_PATH'])
data = loader.load()
print(data[0])
```
